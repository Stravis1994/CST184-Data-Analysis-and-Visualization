```python
import pandas as pd

df = pd.read_csv("renewable_energy.csv")
df.head()

two_columns = df[["Country", "%Renewable"]]
print(two_columns)
```

```python
high_renewable = df[df["%Renewable"] > 70]
high_renewable
```

```python
high_set = set(high_renewable["Country"])
print("Countries with >50% renewable energy:", high_set)
print("Cardinality:", len(high_set))
```

```python
emissions_dict = dict(zip(df["Country"], df["CO2_Emissions"]))
print(emissions_dict["Brazil"])
```

```python
import matplotlib.pyplot as plt

plt.bar(emissions_dict.keys(), emissions_dict.values(), color='green')
plt.xticks(rotation=45)
plt.title("CO₂ Emissions per Capita by Country")
plt.ylabel("Tonnes")
plt.grid(True)
plt.show()
```

```python
plt.scatter(df["Region"], df["%Renewable"])

for i in range(len(df)):
    plt.text(df["Region"][i], df["%Renewable"][i], df['Country'][i])

plt.title("Population vs % Renewable Energy Use")
plt.xlabel("Population (millions)")
plt.ylabel("% Renewable")
plt.grid(True)
plt.show()
```

```python
asia_df = df[df["Region"] == "Asia"]

plt.scatter(asia_df["Population"], asia_df["%Renewable"])

for i in range(len(asia_df)):
    plt.text(
        asia_df["Population"].iloc[i],
        asia_df["%Renewable"].iloc[i],
        asia_df["Country"].iloc[i]
    )

plt.title("Population vs % Renewable Energy Use (Asia)")
plt.xlabel("Population (millions)")
plt.ylabel("% Renewable")
plt.grid(True)
plt.show()
```

```python
subset = df[df["%Renewable"] > 60]
subset.to_csv("high_renewable_countries.csv", index=False)
```
