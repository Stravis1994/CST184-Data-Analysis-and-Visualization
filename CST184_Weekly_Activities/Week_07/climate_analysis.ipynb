```python
import pandas as pd

df = pd.read_csv("owid_climate_data.csv")
df = df.dropna(subset=["CO2 Emissions Per Capita", "Annual Temperature Anomaly"])
df.to_csv('cleaned_owid_climate_data.csv', index=False)
df.head()
```

```python
def get_country_data(country, start_year, end_year):
    country_data = df[df['Country Name'] == country]
    country_data = country_data[country_data['Year'] >= start_year]
    country_data = country_data[country_data['Year'] <= end_year]
    return country_data

get_country_data('India', 2000, 2020)
```

```python
filtered = df[(df['Year'] == 2020) & (df['CO2 Emissions Per Capita'] > 10)][['Country Name', 'CO2 Emissions Per Capita', 'Annual Temperature Anomaly']]
filtered
```
