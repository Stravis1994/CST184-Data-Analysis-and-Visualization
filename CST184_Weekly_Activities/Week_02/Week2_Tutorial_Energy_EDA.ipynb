```python
import pandas as pd

df = pd.read_csv("energy_cleaned.csv")
df.info()
df.describe()

#Question: Are there any unusual values or column names?
#Answer: No the data is clean.
```

```python
high_renew = df[df["%Renewable"] > 50]
print("Countries with >50% renewable energy:", len(high_renew))
```

```python
high_renew = df[df["%Renewable"] > 60]
print("Countries with >60% renewable energy:", len(high_renew))
```

```python
high_renew = df[df["%Renewable"] > 70]
print("Countries with >70% renewable energy:", len(high_renew))

#Question: What proportion of countries use mostly renewable energy?
#Answer: From this data we can see that 2 countries use mostly renewable energy.
```

```python
region_avg = df.groupby("Region")["%Renewable"].mean().plot(kind='bar')
print(region_avg)


# Question: Which region has the highest average renewable use?
#Answer: South America.
```

```python
import matplotlib.pyplot as plt

df["%Renewable"].plot(kind="hist", bins=8, edgecolor="black")
plt.title("Histogram of Renewable Energy Use")
plt.xlabel("% Renewable")
plt.grid(True)
plt.show()

#Question: Are most countries clustered around certain values?
#Answer: Most countries are clustered between 43% - 50% and 90%.
```

```python
import matplotlib.pyplot as plt

df["%Renewable"].plot(kind="hist", bins=5, edgecolor="black")
plt.title("Histogram of Renewable Energy Use")
plt.xlabel("% Renewable")
plt.grid(True)
plt.show()
```

```python
df["%Renewable"].plot(kind="density", linewidth=2)
plt.title("Density Plot of Renewable %")
plt.xlabel("% Renewable")
plt.grid(True)
plt.show()

# Question: Does the distribution appear normal, skewed, or multimodal?
# Answer: The distrubition is skewered to the right.
```

```python
for region, group in df.groupby("Region"):
    if len(group) >= 2:
        group["%Renewable"].plot(kind="density", label=region)
plt.title("Renewable Energy by Region")
plt.xlabel("% Renewable")
plt.grid(True)
plt.show()

# Question: Which region has the highest consistency in renewable use?
# Answer: Asia is the region with the highest consistency in renewable use.
```

```python
for region, group in df.groupby("Region"):
    if len(group) >= 2:
        group["%Renewable"].plot(kind="hist", label=region)
plt.title("Renewable Energy by Region")
plt.xlabel("% Renewable")
plt.grid(True)
plt.show()
```
